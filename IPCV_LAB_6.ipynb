```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

# Histogram streching

```python
lcimg = cv2.imread('/content/a-low-contrast-image-b-after-enhancement_Q320.jpg',0)
```

```python
import numpy as np
nstreched = np.zeros(lcimg.shape)
h1,w1 = nstreched.shape
```

```python
rmax=lcimg.max()
rmin=lcimg.min()
#assuming smin =0 and smax=255
c=255/(rmax-rmin)
m, n = lcimg.shape
```

```python
m,n
```

```python
for i in range(h1):
  for j in range(w1):
    r = lcimg[i][j]
    nstreched[i][j] = int(c*(r-rmin))
```

```python
cv2_imshow(lcimg)
```

```python
cv2_imshow(nstreched)
```

```python
ir_max = lcimg.max()
ir_min = lcimg.min() 
print(ir_max,ir_min)
is_max = nstreched.max()
is_min = nstreched.min() 
print(is_max,is_min)
```

```python
def hist_plot(img):
     
    # empty list to store the count
    # of each intensity value
    count =[]
     
    # empty list to store intensity
    # value
    r = []
     
    # loop to traverse each intensity
    # value
    for k in range(0, 256):
        r.append(k)
        count1 = 0
         
        # loops to traverse each pixel in
        # the image
        for i in range(m):
            for j in range(n):
                if int(img[i, j])== k:
                    count1+= 1
        count.append(count1)
         
    return (r, count)
 
```

```python
r1, count1 = hist_plot(lcimg)
 
# plotting the histogram
plt.stem(r1, count1)
plt.xlabel('intensity value')
plt.ylabel('number of pixels')
plt.title('Histogram of the original image')
```

```python
r1, count1 = hist_plot(nstreched)
 
# plotting the histogram
plt.stem(r1, count1)
plt.xlabel('intensity value')
plt.ylabel('number of pixels')
plt.title('Histogram of the stretched image')
```
