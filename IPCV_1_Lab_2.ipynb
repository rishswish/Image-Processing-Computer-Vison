```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('/content/TAJ MAHAL.jpg')
cv2_imshow(img)
```

**1] Digital Negative**

```python
row , col , _ = img.shape
```

```python
for i in range(0,row-1):
  for j in range(0,col-1):
    pixel = img[i,j]
    pixel[0] = 255 - pixel[0]
    pixel[1] = 255 - pixel[1]
    pixel[2] = 255 - pixel[2]

    img[i,j] = pixel
```

```python
cv2_imshow(img)
```

**2] Thresholding**

```python
img = cv2.imread('/content/TAJ MAHAL.jpg',0)
```

```python
threshold = 150
```

```python
for i in range(0,row-1):
  for j in range(0,col-1):
    pixel = img[i,j]

    if(pixel<threshold):
      pixel = 0
    else:
      pixel = 255


    img[i,j] = pixel
```

```python
cv2_imshow(img)
```

**3] Gray Level Slicing**

```python
a = 100
b = 150
```

a] Without background

```python
img = cv2.imread('/content/TAJ MAHAL.jpg')
```

```python
for i in range(0,row-1):
  for j in range(0,col-1):
    pixel = img[i,j]

    if(a<=pixel[0]<=b):
      pixel[0]=255
    else:
      pixel[0]=0

    if(a<=pixel[1]<=b):
      pixel[1]=255
    else:
      pixel[1]=0

    if(a<=pixel[2]<=b):
      pixel[2]=255
    else:
      pixel[2]=0

    img[i,j] = pixel
```

```python
cv2_imshow(img)
```

b] With background

```python
img = cv2.imread('/content/TAJ MAHAL.jpg')
```

```python
for i in range(0,row-1):
  for j in range(0,col-1):
    pixel = img[i,j]

    if(a<=pixel[0]<=b):
      pixel[0]=255

    if(a<=pixel[1]<=b):
      pixel[1]=255

    if(a<=pixel[2]<=b):
      pixel[2]=255

    img[i,j] = pixel
```

```python
cv2_imshow(img)
```
