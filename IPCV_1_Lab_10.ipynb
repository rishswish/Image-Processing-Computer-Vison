```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
import numpy as np
import random
```

```python
def display(img):
  fig=plt.figure(figsize=(5,6))
  ax=fig.add_subplot(111)
  ax.imshow(img,cmap='gray')
  ax.axis('off')

def create_img():
  blank_img=np.zeros((200,200))
  font=cv2.FONT_HERSHEY_DUPLEX
  cv2.putText(img=blank_img,
              text='H',
              org=(50,150),
              fontFace=font,
              fontScale=5,
              color=(255,255,255),
              thickness=25,
              lineType=cv2.LINE_AA)
  return blank_img
```

# **Sobel Filter**

```python
sobel_x=np.array([[-1,-2,-1],
                  [0,0,0],
                  [1,2,1]])

sobel_y=np.array([[-1,0,1],
                  [-2,0,2],
                  [-1,0,1]])
```

```python
i=create_img()
display(i)
```

```python
edge_x=cv2.filter2D(src=i,ddepth=-1,kernel=sobel_x)
display(edge_x)
```

```python
edge_x[edge_x!=0]=255
display(edge_x)
```

```python
edge_y=cv2.filter2D(src=i,ddepth=-1,kernel=sobel_y)
display(edge_y)
```

```python
edge_y[edge_y!=0]=255
display(edge_y)
```

```python
edge=edge_x+edge_y
```

```python
edge[edge!=0]=255
display(edge)
```

```python
edge_x[edge_x!=0]=255
display(edge_x)
```

```python
edge_y=cv2.filter2D(src=i,ddepth=-1,kernel=pw_y)
display(edge_y)
```

```python
edge_y[edge_y!=0]=255
display(edge_y)
```

```python
edge=edge_x+edge_y
edge[edge!=0]=255
display(edge)
```
